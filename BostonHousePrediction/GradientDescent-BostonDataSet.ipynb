# Hardcoding Gradient Descent

```python
def step_gradient(X,Y,m,learning_rate):
    m_slope = np.zeros(len(X[0]))
    N = len(X)
    for i in range(N):
        x = X[i]
        y = Y[i]
        for j in range(len(x)):
            m_slope[j]+=(-2/N)*(y-sum(m*x))*x[j]
    new_m = m - learning_rate*m_slope
    return new_m
```

```python
def cost(m,x,y):
    total_cost = 0
    N = len(x)
    for i in range(N):
        total_cost += (1/N)*((y[i]-sum(m*x[i]))**2)
    return total_cost 
    
```

```python
def gd(x,y,learning_rate,num_iterations):
    m = np.zeros(len(x[0]))
    for i in range(num_iterations):
        m = step_gradient(x,y,m,learning_rate)
        print(i," Cost: ",cost(m,x,y))
    return m
```

```python
def run(x,y):
    learning_rate = 0.1
    num_iterations = 300
    m = gd(x,y,learning_rate,num_iterations)
    return m
```

```python
x=training[:, :-1]
y=training[:, -1]
run(x,y)
```

```python
x=training[:, :-1]
y=training[:, -1]

m=gradient_descent(x, y)
```
